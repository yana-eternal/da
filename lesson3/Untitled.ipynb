```python
def split_brand (brand_name_data):
    return brand_name_data.split(' ') [-1]
```

```python
user_df['brand_name'] = user_df.brand_info.apply(split_brand) #добавили колонку в начале
```

```python
user_df.head ()
```

```python
user_df['brand_name'] = user_df.brand_info.apply(lambda x: x.split(' ') [-1])
#более компкатный способ получить имя брэнда.используем lambda функцию чтобы разрезать строку по пробелу и взять последний элемент
```

```python
users_purchases = ( user_df
                   .groupby('user_id', as_index = False)
                   .agg ({'brand_name':'count'}) 
                   .rename (columns = {'brand_name':'purchases'})
                  )
users_purchases
```

```python
users_purchases.purchases.median()
```

```python
users_purchases.describe ()
#75% больше 5 покупок
#50% медиана.половина пользователей купили больше 2 покупоке
```

```python
users_purchases = ( user_df
                   .groupby('user_id', as_index = False)
                   .agg ({'brand_name':'count'}) 
                   .rename (columns = {'brand_name':'purchases'})
                   .query ('purchases>=5'))
users_purchases
```

```python
lovely_brand_purchases_df = user_df.groupby(['user_id','brand_name' ], as_index = False)\
        .agg ({'brand_info':'count'}) \
        .sort_values(['user_id','brand_info'], ascending=[False, False]) \
        .groupby ('user_id') \
        .head(1) \
        .rename (columns={'brand_name':'lovely_brand', 'brand_info':'lovely_brand_purchases'})
lovely_brand_purchases_df      
```

```python
user_un_br = user_df.groupby ('user_id',as_index = False) \
        .agg({'brand_name':pd.Series.nunique}) \
        .rename (columns={'brand_name':'unique_brands'})
user_un_br.head(5)
```

```python
users_purchases = ( user_df
                   .groupby('user_id', as_index = False)
                   .agg ({'brand_name':'count'}) 
                   .rename (columns = {'brand_name':'purchases'})
                   .query ('purchases>=5'))
users_purchases
```

```python
user_un_br = user_df.groupby ('user_id',as_index = False) \
        .agg({'brand_name':pd.Series.nunique}) \
        .rename (columns={'brand_name':'unique_brands'})
user_un_br.head(5)
```

```python
lovely_brand_purchases_df = user_df.groupby(['user_id','brand_name' ], as_index = False)\
        .agg ({'brand_info':'count'}) \
        .sort_values(['user_id','brand_info'], ascending=[False, False]) \
        .groupby ('user_id') \
        .head(1) \
        .rename (columns={'brand_name':'lovely_brand', 'brand_info':'lovely_brand_purchases'})
lovely_brand_purchases_df 
```

```python
loyalty_df = users_purchases\
        .merge (user_un_br, on='user_id') \
        .merge (lovely_brand_purchases_df, on='user_id') 
loyalty_df.head(5)
```

```python
loyal_users = loyalty_df[loyalty_df.unique_brands==1]
```

```python
loyalty_df['loyalty_score'] = loyalty_df.lovely_brand_purchases / loyalty_df.purchases
```
